```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

airbus = "C:/Users/anant/OneDrive/Desktop/ENAC IATOM/Sem 7/Economics/Project/BA_AIR_data_new.csv"
df = pd.read_csv(airbus)
df.tail()
```

```python
df.info()
```

```python
column_name = '%Change_BA'

# Calculate the IQR (Interquartile Range)
Q1 = df[column_name].quantile(0.25)
Q3 = df[column_name].quantile(0.75)
IQR = Q3 - Q1

# Define a threshold to identify outliers
threshold = 1.5

# Find values outside the range [Q1 - threshold * IQR, Q3 + threshold * IQR]
outliers = df[(df[column_name] < Q1 - threshold * IQR) | (df[column_name] > Q3 + threshold * IQR)]

# Display the outliers
print("Outliers:")
print(outliers)
```

```python

outliers = outliers[outliers.columns[[0,20,21,22,23,24,11,17,18,19,9,15,18,19]]]
outliers
```

```python
outliers['Count'] = outliers.reset_index().index + 1
outliers
```

```python
g = sns.pairplot(outliers, palette="husl", diag_kind="kde")
```

```python
correlation = outliers.corr()
plt.figure(figsize=(18, 10))
sns.heatmap(correlation, xticklabels = correlation.columns, yticklabels = correlation.columns,cmap=sns.diverging_palette(220, 10, as_cmap=True), annot=True)
```
